```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("IRIS.csv")

# Quick look
print(df.head())
print(df.info())

```

1. Bar Plot – Count of species

```python
plt.figure(figsize=(6,4))
sns.countplot(x="species", data=df)
plt.title("Number of Samples per Species")
plt.show()

```

2. Box Plot – Sepal length by species

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="species", y="sepal_length", data=df)
plt.title("Sepal Length Distribution by Species")
plt.show()
```

3. KDE Plot – Petal length distribution

```python
plt.figure(figsize=(6,4))
sns.kdeplot(df["petal_length"], shade=True)
plt.title("KDE Plot of Petal Length")
plt.show()
```

4. Violin Plot – Petal width by species

```python
plt.figure(figsize=(6,4))
sns.violinplot(x="species", y="petal_width", data=df)
plt.title("Petal Width Distribution by Species")
plt.show()
```

5. Swarm Plot – Sepal width by species

```python
plt.figure(figsize=(6,4))
sns.swarmplot(x="species", y="sepal_width", data=df)
plt.title("Sepal Width Values per Species")
plt.show()
```
